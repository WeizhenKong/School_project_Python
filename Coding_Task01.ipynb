#### Using yob1880.txt and yob1980.txt, create a list of lists, where the count of names in more than 100 but less than or equal to 5000

```python
# Read in yob1880
mydf = pd.read_csv('names/yob1880.txt', header=None, names=['name','gender','count'])
mydf.shape
```

```python
# Read in yob1980
mydf2 = pd.read_csv('names/yob1980.txt', header=None, names=['name','gender','count'])
mydf2.shape
```

```python
# Append mydf2 to mydf
df = mydf.append(mydf2)
mydf.shape
```

```python
# Combine the two dataset to form a single list of lists
list_of_lists = [[name[0],name[1],name[2]] for name in mydf.values]
list_of_lists
```

```python
# Filter for where count is less than or equal to 5000 and more than 100
myresult = [name[0] for name in list_of_lists if name[2] > 100 and name[2]<= 5000]
print(len(myresult))
myresult
```

#### Using lists (or list of lists) find the average number of births for names that start with the letter A

```python
# Using lists
myresult = [name[2] for name in list_of_lists if name[0].startswith('A')]
pd.Series(myresult).mean()
```

```python
# Using pandas
mydf = pd.DataFrame(list_of_lists, columns=['name','gender','count'])
mydf.loc[mydf['name'].str.startswith('A'), 'count'].mean()
```

#### Using lists (or list of lists) find the average number of births that end with a

```python
# Using lists
myresult = [name[2] for name in list_of_lists if name[0][-1::-1].startswith('a')]
pd.Series(myresult).mean()
```

```python
# Using pandas
myendsa = []
for i in range(0,len(mydf)):
    if mydf['name'][i][-1::-1].startswith('a'):
        myendsa.append(mydf['count'][i])
    
pd.Series(myendsa).mean()
```

#### Create a dictionary named d1880 for all names for the yob1880.txt file where the key is the name and the value is the sex and frequency pair

```python
# Using lists

# Import yob1880.txt data into list of lists
mydf = pd.read_csv('names/yob1880.txt', header=None, names=['name','gender','count'])
list_of_lists = [[name[0],name[1],name[2]] for name in mydf.values]

# Create dictionary from list of lists
key = [name[0] for name in list_of_lists]
value = [[name[1],name[2]] for name in list_of_lists]
d1880 = dict(zip(key, value))

# Check first value
d1880['Anna']
```
